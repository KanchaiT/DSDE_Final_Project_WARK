```python
import pandas as pd
import numpy as np
from itertools import combinations
```

```python
df = pd.read_csv('joined_2018-2023.csv')
```

```python
df.head()
```

```python
split_author = df['authors'].str.split(';')
df['authors'] = split_author
```

```python
col = ['citation_title','authors']
df[col]
```

```python
def extract_authors(authors_string):
  """Extracts a list of authors from a string.

  Args:
    authors_string: A string containing a list of authors in the format
      "[Author 1, Author 2, ...]".

  Returns:
    A list of author names.
  """
  authors_string = authors_string.strip('[]')  # Remove brackets
  authors_list = [author.strip() for author in authors_string.split(',')]  # Split and strip spaces
  return authors_list
```

```python
pairings = []
for index, row in df.iterrows():
    authors_list = row['authors'] # Access the list directly
    if len(authors_list) > 1:
        for pair in combinations(authors_list, 2):
            pairings.append({'citation_title': row['citation_title'], 'author1': pair[0], 'author2': pair[1]})
    elif len(authors_list) == 1:
        pairings.append({'citation_title': row['citation_title'], 'author1': authors_list[0], 'author2': authors_list[0]})

pairing_df = pd.DataFrame(pairings)

print(pairing_df)
```
